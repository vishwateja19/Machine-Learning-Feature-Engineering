```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("dataset_1.csv")
```

```python
df.head()
```

```python
X = df.iloc[: , :-1]
y = df.iloc[: , -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from feature_engine.selection import DropConstantFeatures
```

```python
reg = DropConstantFeatures(tol=1,variables = None , missing_values='raise')
```

```python
reg.fit(X_train)
```

```python
reg.features_to_drop_
```

```python
reg.n_features_in_
```

```python
len(reg.features_to_drop_)
```

```python
X_train = reg.transform(X_train)
```

```python
X_test = reg.transform(X_test)
```

```python
X_train.head()
```
