### constant feartures means each observation has only same value and their variance will be = 0 , so we can remove those values definatelly 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("dataset_1.csv")
```

```python
df.head()
```

```python
for i in df.columns:
    print(i , 'Null values :',df[i].isnull().sum())
```

### constant features removal 

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
reg = VarianceThreshold()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[: , :-1]
y = df.iloc[: , -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape , X_test.shape
```

```python
reg.fit(X_train)
```

```python
sum(reg.get_support())   # 258 feturess are not constant 
```

```python
X_train.columns[reg.get_support()]
```

```python
constant = X_train.columns[~reg.get_support()]

len(constant)
```
