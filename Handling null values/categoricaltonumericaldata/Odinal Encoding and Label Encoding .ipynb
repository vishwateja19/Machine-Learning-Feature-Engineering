```python
X = df.iloc[: , :-1]
y = df.iloc[: ,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
a = reg.transform(X_train[['review']])
```

```python
data = pd.DataFrame(a)
```
